```
import pandas as pd
df=pd.read_csv("Sales.csv", sep=";")
print(df.columns)
print(df.head(0))
print(df.tail(1))
print(df.index)
print(df[df["Product Weight"]<15])
df.iloc[:-2]
df.loc[4]
print(df[df["Order Quantity"]>5])
print(df[df["Order Quantity"]<2])
print(df[df["Product Price"]<800])
print(df[df["Product Price"]>1700])
print(df[df["Order Total"]>5000]) 
print(df[(df["Order Total"]>5000) & (df["Payment Method"]=="Credit Card")])
print(df[(df["Order Total"]<1000) & (df["Order Quantity"]<2)])
print(df[(df["Order Total"]>5000) | (df["Payment Method"]=="PayPal")])
print(df[(df["Product Price"]>5000)| (df["Order Total"]>1000)])
print(df[(df["Product Price"]<5000) & (df["Payment Method"]=="Credit Card")])
print(df["Product Category"].unique())
print(df[(df["Product Category"].isin(
    ["Mountain Bike", "Mountain Bike XC"])) & (df["Product Price"]>1000) & (df["Order Quantity"]>1)])

top_revenue=df.loc[df["Order Total"].idxmax(),["Product Name","Order Total"]]
print("Najvyšší tržby:\n", top_revenue)
low_revenue=df.loc[df["Order Total"].idxmin(),["Product Name","Order Total"]]
print("Nejnižší tržby:\n", low_revenue)

top_sales=df.loc[df["Order Quantity"].idxmax(),["Product Name","Order Quantity"]]
print("Najvyšší prodej:\n", top_sales)
low_sales=df.loc[df["Order Quantity"].idxmin(),["Product Name","Order Quantity"]]
print("Nejnižší prodej:\n", low_sales)

import matplotlib.pyplot as plt
products=[top_revenue["Product Name"], low_revenue["Product Name"]]
values=[top_revenue["Order Total"], low_revenue["Order Total"]]
colors=["seagreen", "salmon"]
plt.bar(products, values, color=colors)
plt.title("Nejvyšší vs. Nejnižší tržby")
for i, v in enumerate(values):
    plt.text(i, v+100,str(v), ha="center", fontsize=10, fontweight="bold")
plt.show()

products=[top_sales["Product Name"], low_sales["Product Name"]]
values=[top_sales["Order Quantity"], low_sales["Order Quantity"]]
colors=["green", "orange"]
plt.bar(products, values, color=colors)
plt.title("Nejvyšší vs. Nejnižší prodej")
for i, v in enumerate(values):
    plt.text(i, v+100,str(v), ha="center", fontsize=10, fontweight="bold")
plt.show()
```
